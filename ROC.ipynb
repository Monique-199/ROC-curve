```python
import matplotlib.pyplot as plt
```

```python
# FPR and TPR values for Algorithm A
fpr_a = [0, 0.01, 0.19, 0.58, 0.98, 1]
tpr_a = [0, 0.78, 0.91, 0.96, 1, 1]
```

```python
# FPR and TPR values for Algorithm B
fpr_b = [0, 0.01, 0.2, 0.32, 0.96, 1]
tpr_b = [0, 0.42, 0.58, 0.64, 0.9, 1]
```

```python
plt.figure(figsize=(8, 6))
plt.plot(fpr_a, tpr_a, label='Algorithm A')
plt.plot(fpr_b, tpr_b, label='Algorithm B')
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curves for Algorithms A and B')
plt.legend()
plt.grid()
plt.show()
```
